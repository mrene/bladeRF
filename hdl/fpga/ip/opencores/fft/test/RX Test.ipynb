```python
def getFFTSamples(src, fftlen):
    data = list(itertools.islice(src, 0, fftlen))
    data = np.divide(data, float((2**16)))
    #data = np.fft.fft(data)
    yield data
    
def computeMag(src, fftlen):
    pwr_scale = 1.0 / (fftlen**2);
    
    for x in src:
        yield np.multiply(np.abs(x)**2, pwr_scale)
        
def computeDB(src):
    for x in src:
        val = np.add(x, 1e-20)
        val = 10 * np.log10(x)
        yield val

def computeAverage(src, n):
    data = list(itertools.islice(src, 0, n))
    return np.average(data, axis=0)

def computeAverage2(src, fftlen, n = 100, avg = 0.75):
    src = list(itertools.islice(src, 0, n))
    data = np.ones(fftlen) * -120
    count = 0
    for bins in src:
        if count == n:
            return data
        data = np.add( np.multiply(data, 1-avg), np.multiply(bins, avg) )
        count += 1
    return data
```

```python
fftlen = 2**13
samp_rate = 20e6
center_freq = 540e6
freq = np.fft.fftfreq(fftlen, 1./samp_rate)
freq = np.add(freq, center_freq)

pwr_scale = 1.0 / (fftlen**2)

def setupSrc():
    return binaryfile('./data-rxfft.bin', samplesize=32)

data = setupSrc()
data = getFFTSamples(data, fftlen)
data = computeMag(data, fftlen)
data = computeDB(data)
data = computeAverage2(data, fftlen, n=int(40e6*2), avg=0.75)
data = data
plt.figure(figsize=(13, 5))
plt.ticklabel_format(useOffset=False,style='plain')
plt.ylim([-120,-20])
plt.plot(freq, data); None
```
